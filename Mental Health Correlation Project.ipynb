```python
# Import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

# Read in the data

df = pd.read_csv(r'C:\Users\zahas\Documents\Downloads\MentalHealthDataset.csv') 
```

```python
# let's look at the data

df.head()
```

```python
# Drop any duplicates

df.drop_duplicates()
```

```python
# Find any missing data

for col in df.columns:
    perct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, perct_missing))
    
```

```python
# Check for missing values in self_employed column

df.loc[:, "self_employed"]
```

```python
# Convert NaN values to empty

df.fillna('', inplace =True)
```

```python
# Create a year column from Timestamp column

df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df['Year'] = df['Timestamp'].dt.year
```

```python
# Converting each unique object to a numerical value

df_numerical = df

for col_name in df_numerical.columns:
    if(df_numerical[col_name].dtype == 'object'):
        df_numerical[col_name] = df_numerical[col_name].astype('category')
        df_numerical[col_name] = df_numerical[col_name].cat.codes

df_numerical
```

```python
# Correlation matrix between all numerical values

correlation_matrix = df_numerical.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matric') 

plt.xlabel('Data label')

plt.ylabel('Data label')

plt.rcParams['figure.figsize'] = (20,10)

plt.show()
```

```python
# Correlation matrix values as a table

df_numerical.corr()
```

```python
# Organize to check for high correlation by unstacking each columns

corr_matrix = df_numerical.corr()

corr_pairs = corr_matrix.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
# Correlation matrix above 0.3

high_corr = sorted_pairs[(sorted_pairs) > 0.3]

high_corr
```
